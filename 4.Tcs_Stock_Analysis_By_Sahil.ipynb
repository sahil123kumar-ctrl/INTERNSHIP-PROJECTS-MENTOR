# 9. Distribution of Tcs Stock Split over years

A stock split is when a company divides its existing shares into multiple new shares to boost the stock's liquidity.
The total value of the company doesn't change, but the number of shares increases,
and the price per share decreases proportionally.

```python
df.columns
```

```python

# Step 1: Group by year and get the sum of stock splits 
stock_splits_yearly = df.groupby('Year')['Stock Splits'].sum().reset_index()

# Step 2: Filter out years where stock splits actually occurred 
stock_splits_yearly = stock_splits_yearly[stock_splits_yearly['Stock Splits'] > 0]

# Step 3: Plotting
plt.figure(figsize=(8, 5))
plt.plot(stock_splits_yearly['Year'], stock_splits_yearly['Stock Splits'], 
         marker='s', linestyle='-', color='purple', linewidth=2)

plt.title(" TCS Stock Splits Over the Years", fontsize=15, weight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Stock Split Ratio", fontsize=12)
plt.xticks(stock_splits_yearly['Year'], rotation=45, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

The line graph shows that the TCS stock split ratio remained constant at 2.00 from 2006 to 2018. 
This indicates that TCS has not changed its stock split ratio during these years, reflecting a stable approach to stock split policy over this period.

# 10. Stock Prices Trend Distibution:

Creating  a new column 'Trend' 

1.'Up' if Close > Open

2.'Down' if Close < Open

3.'No Change' otherwise

```python
df.loc[df['Close'] > df['Open'],'Trend'] = 'Up'
df.loc[df['Close'] < df['Open'],'Trend'] = 'Down'
df.loc[df['Close'] == df['Open'],'Trend'] = 'No Change'
```

```python
df['Trend'].value_counts()
```

```python
# Step 1: Get the value counts of the 'Trend' column
trend_counts = df['Trend'].value_counts().reset_index()
trend_counts.columns = ['Trend', 'Count']

# Step 2: Plotting
sns.barplot(x='Trend', y='Count', data=trend_counts, palette='viridis',hue='Trend')

# 3.Adding labels and title
plt.title('Trend Distribution', fontsize=14)
plt.xlabel('Trend', fontsize=12)
plt.ylabel('Number of Occurrences', fontsize=12)

for index, row in trend_counts.iterrows():
    plt.text(x=index, y=row['Count'] + 1, s=row['Count'], ha='center', fontsize=11)

plt.tight_layout()
plt.show()
```

# Insight:

Trend Distribution Summary
Most Frequent:
"Down" trend dominates with 2,330 occurrences â€” indicates frequent declines.

Second Highest:
"Up" trend appears 2,097 times â€” shows many upward movements.

Least Common:
"No Change" has only 36 instances â€” stability is rare.

# 13. No Movement Day:

 IF ON A DAY OPEN PRICE = HIGH PRICE = lOW PRICE = CLOSE PRICE THAT DAY IS CALLED  NO Movement day':

```python
# 12.For How many days it was the NoMovementDay:

df['No_Movement_Day'] = df.apply(lambda x :'Yes' if x['Open'] == x['High'] == x['Low'] == x['Close'] else 'No' , axis = 1)

df['No_Movement_Day'].value_counts()
```

```python
# Step 1: Count values
no_movement_counts = df['No_Movement_Day'].value_counts().reset_index()
no_movement_counts.columns = ['No_Movement_Day', 'Count']

# Step 2: Plot
plt.figure(figsize=(6, 4))
sns.barplot(x='No_Movement_Day', y='Count', data=no_movement_counts, palette='pastel',hue='No_Movement_Day')

# Step 3: Add labels and title
plt.title('No Movement Day Distribution', fontsize=14)
plt.xlabel('No Movement Day', fontsize=12)
plt.ylabel('Number of Days', fontsize=12)


for index, row in no_movement_counts.iterrows():
    plt.text(index, row['Count'] + 1, row['Count'], ha='center', fontsize=11)

plt.tight_layout()
plt.show()
```

There are 25 days when the stock price remained constant throughout the day

Meaning:

A stock is said be highly volatile if intradity volatidity is between 3 to 5 percent:

intraday_volatility_% = ((high-low)/open)) * 100

```python
df['Intraday_Volatility_%'] = ((df['High'] - df['Low']) / df['Open']) * 100
```

if'Intraday_Volatility_% <= 1.5 then "Low_Volatile"

if 'Intraday_Volatility_% >1.5 and <=3.5 then "Volatile

```python
df.loc[df['Intraday_Volatility_%'] <= 1.5, 'Volatile_Level'] = 'Low_Volatile'
df.loc[(df['Intraday_Volatility_%'] > 1.5) & (df['Intraday_Volatility_%'] <= 3.5), 'Volatile_Level'] = 'Volatile'
df.loc[df['Intraday_Volatility_%'] > 3.5, 'Volatile_Level'] = 'High_Volatile'
```

# 20. DATES with the continous increasing stock prices streaks:

```python
# Step 1: Calculate if price is increasing
df['Prev_Close'] = df['Close'].shift(1)
df['Increasing'] = df['Close'] > df['Prev_Close']

# Step 2: Identify streaks
df['Streak_ID'] = (~df['Increasing']).cumsum()

# Step 3: Filter only increasing rows
streaks = df[df['Increasing']].copy()

# Step 4: Find the longest streak
longest_streak = streaks.groupby('Streak_ID').size().idxmax()
streak_data = streaks[streaks['Streak_ID'] == longest_streak]

print(" Maximum Continuous Increasing Days:", len(streak_data))
print(" Start Date:", streak_data['Date'].iloc[0])
print(" End Date:", streak_data['Date'].iloc[-1])
print("\n Price Data:")
print(streak_data[['Date', 'Close']])
```

# Reason:

and renewed investor confidence post-COVID disruptions

# 21.Dates with the longest continous Decrease stock Prices streak:

```python
# Step 1: Calculate Previous Close
df['Prev_Close'] = df['Close'].shift(1)

# Step 2: Mark Decreasing Days
df['Decreasing'] = df['Close'] < df['Prev_Close']

# Step 3: Create group IDs for each continuous decreasing streak
df['Streak_ID_Decreasing'] = (~df['Decreasing']).cumsum()

# Step 4: Filter decreasing rows
decrease_streaks = df[df['Decreasing']].copy()

# Step 5: Find the longest decreasing streak
longest_decrease = decrease_streaks.groupby('Streak_ID_Decreasing').size().idxmax()
decreasing_data = decrease_streaks[decrease_streaks['Streak_ID_Decreasing'] == longest_decrease]

print(" Maximum Continuous Decreasing Days:", len(decreasing_data))
print(" Start Date:", decreasing_data['Date'].iloc[0])
print(" End Date:", decreasing_data['Date'].iloc[-1])
print("\n Price Data:")
print(decreasing_data[['Date', 'Close']])
```
